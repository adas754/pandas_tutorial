```python
# how will create a dataframe from dictionary
import pandas as pd

d={'name':['a','b','c','d','x','y','z'],
   "roll_no":[1,2,3,4,5,6,7],
   "score":[10,20,30,40,50,60,70]}
```

```python
df=pd.DataFrame(d)
df
```
